# Request Response Cycle



```python
# Its very important to understand the request response cycle 
# We want to scrap data from web, so we need to understand the basics of web 
# What happens if we google search about something, we get a webpage in return, right
# Its simple enough till this point but there's a lot that happens in the backend

# I'll explain in very simple words, so I might be a little off in my explaination
# But my intention here is to give an intuition about how internet works

# When we go to a website, we either get a webpage as an output on our screen
# Or we get an error (error 404 page does not exist)

# Simple enough, right? But there's a lot that happens in the backend though

# Let's understand a server first
# A server is just another computer owned by someone else, which is connected to internet
# And just because we're using windows on our machines
# doesn't mean, the other person is also using windows, right
# So we have web protocols like https
# https means hypertext transfer protocol secure
# www means world wide web
# https says we all should converse in the same language over the web
# lets take an example

# First let's say we go to https://www.google.com/search?q=apple

# We request the server "google.com" over the world wide web
# using the https protocol that
# We want to access your server and 
# we want to search apple over the web using your server

# The search?q=apple argument in the url means its a dictionary in an url format
# a dictionary is a data structure which has key:value pairs  
# key of the dictionary is "search", and the value is user defined 
# In our case the value is "apple"
# In python this dictionary would look like {"search":"apple"}

# So, we make a request over the world wide web
# that we want to access "google.com" server
# and using that server, we want to search the word "apple" using the
# https protocol

# Once our request is accepted by the server "google.com"
# It returns a response object first
# On the web, the response is usually written in html 
# So, we need to convert that response to html

# So, in a nutshell, this was request response cycle 
# We send a request and get a response in return

# A little extra info
# Character encoding: 
# People who use english, type characters in their keyboards like
# the letter from a to z, from capitalized A to Z
# the numerics 0-9
# and the special character like $ or !
# This is all a part of ascii
# You can google ascii for more info 
# But say a person who does not speak english
# how would he interpret these english letters?
# a japanese person use a keyboard with japanese letters
# So, we need a way to convert our english language to japanese in
# order to communicate with folks around the world who don't
# speak english

# In the same way, different machines have different
# characters that they can understand
# python works on I guess unicode
# But the world wide web has I guess utf-8 character encoding

# So, some times, we need to be careful to choose the right character encoding as well
# So, there are times we need to encode or decode our character encodings properly

# Our modern day browsers automate so much work behind the scenes, right :)

```

# Better way of parsing data and getting rid of unwanted characters

```python
# Lets first understand the problem
# we have a price column which has a weird currency character in it
# because of which, we cant directly convert our price to integers
# As part of any analysis, we need to convert price from string to integers or floats
# Say, to find the minimum price, we can use minimum only on numeric data

# So, we need to pull our numeric digits out of the price string
# For that we'll use additional functionality known as
# Regular Expression
# import re # this is the module for regular expressions
```

```python
import re

# Lets say we had a string out of which we need to pull out
# a numeric data
string1 = "@#$%$@#342523452435"

# First, lets see the re module
# We'll soon discuss the pattern to extract don't worry
pattern_to_extract = "[0-9]"

what_is_this_variable = re.findall(pattern_to_extract, string1)
print(what_is_this_variable)
```

```python
# first, we use re.findall to find a pattern in data
# the pattern found every numeric data and appended that
# indivisual numeric data to a list

# now that we have our numeric data lets convert it into integers
price_string = "".join(what_is_this_variable)
# this will join our list
print(price_string)
```

```python
# Now we convert this joined string to integer
price_integer = int(price_string)
print(price_integer)
```

```python
# Now, lets talk about the pattern we used to extract numerics
pattern = "[0-9]"
# this means, pull out numbers matching every digit between 0 to 9 
# [0-9] is a syntax of regular expressions
# we could also have used [0123456789] as a pattern
pattern2 = "[0123456789]"
print(re.findall(pattern2, string1))
```

```python
# To understand the pattern matching lets pull out the character
# a from a string " a b ac"
pattern_a = "[a]"
some_string = " a b ac"
print(re.findall(pattern_a, some_string))

# the output will be, find the pattern as many times it 
# occurs in our string
```

```python
# Lets try another example
pattern_5 = "[5]"
some_string = "abs sda 555 666"

# We want to pull the digit 5 out of this string
print(re.findall(pattern_5, some_string))
```

```python
# Now with pattern matching under our belt
# lets convert our dataframe price to integers

# But lets just do one more example first
# Lets convert this string to integer

pattern = "[0123456789]"
some_string = "$56,700,987"

answer = re.findall(pattern, some_string)

print("At this point our variable answer is a list \n", answer)
# the \n just introduce a new line in our console output
# So that our list would be printed in a new line
# So, \n is just a new line character which is not visible in output
print()
answer = "".join(answer)
print("Now our answer variable list has been converted to string \n", answer)
print()
answer = int(answer)
print("our answer variable must now be an integer, \n", answer)
print(type(answer))
```

# Create our parsed dataframe

```python
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")

products = []
prices = []

for a in soup.find_all("a", href=True, attrs={"class":"_1fQZEK"}):
    name = a.find("div", attrs={"class":"_4rR01T"})
    price=a.find("div", attrs={"class":"_30jeq3 _1_WHN1"})
    products.append(name.text)                                                                                                                                       
    prices.append(price.text)

df = pd.DataFrame({"Product Name":products, "Price":prices})

df.head()
```

```python
# We will create a custom function which we will apply 
# row by row on price column of our dataframe
# so on each iteration, our function will convert the
# price in string format to integer format

# We use a default pattern here, which means
# if we don't specify pattern while using this function
# the default value of pattern argument will be "[0123456789]"
# some_string will assume price column string row by row
def convert_data_using_pattern(some_string, pattern="[0123456789]"):
    import re
    our_data = re.findall(pattern, some_string)
    our_data = "".join(our_data)
    our_data = int(our_data)
    return our_data
```

```python
df["Price"]
```

```python
df["Price"].apply(
        lambda column_price: convert_data_using_pattern(column_price)
    ) 
```

```python
df["Price"] = df["Price"].apply(
    lambda column_price: convert_data_using_pattern(column_price))
```
