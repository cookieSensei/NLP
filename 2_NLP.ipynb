# Greedy matching

```python
import re
x = 'From: Using the : character'
y = re.findall('^F.+:', x)
print(y)

# Search till the last :
```

```python
import re
x = 'From: Using the : character'
y = re.findall('^F.+?:', x)
print(y)

# Search till first encounter of :
```

```python
x = 'From stephen.marquard@uct.ac.za Sat Jan  5 09:14:16 2008'
y = re.findall('\S+@\S+', x)
print(y)
```

```python
x = 'From stephen.marquard@uct.ac.za Sat Jan  5 09:14:16 2008'
y = re.findall('\S+?@\S+?', x)
print(y)
```

```python
# start from caret but start sxtraction only from ()
x = 'From stephen.marquard@uct.ac.za Sat Jan  5 09:14:16 2008'
y = re.findall('^From (\S+@\S+)', x)
print(y)
```

```python
# [^ ] says: anything but a whitespace, same as \S
# inside the square brackets ^ acts as a "not"
# like [^ ] means: not a whitespace

import re
line = 'From stephen.marquard@uct.ac.za Sat Jan  5 09:14:16 2008'
y = re.findall('[^ ]@([^ ]+)', line)
print(y)
```

```python
# [^ ] says: anything but a whitespace, same as \S
# inside the square brackets ^ acts as a "not"
# like [^ ] means: not a whitespace
# So, [^F] would mean: not F

import re
line = 'From stephen.marquard@uct.ac.za Sat Jan  5 09:14:16 2008'
y = re.findall('[^F]', line)
print(y)

```

```python
import re
line = 'From stephen.marquard@uct.ac.za Sat Jan  5 09:14:16 2008'
y = re.findall('.*@([^ ]+)', line)
print(y)
```

```python
# $ is a special character
# To look for a dollar sign in a string 
# we'd need to use \$
# \ back slash is used to escape characters
# like \n means new line
```

# Using only Language Processing

```python
url = "https://confrecordings.ams3.digitaloceanspaces.com/Reviews.csv"
df = pd.read_csv(url)


#break into sentences
df["sentence"]=df["Tweets"].str.split(".")
#break into words
df["words"]=df["Tweets"].str.split()
#count frequency
df["Tweets"].str.split(expand=True).stack().value_counts()
#remove stopwords
word_stop=["is","are","you","so","a","in","up","the","as","it","I","can","the","did","an","You","it's","of","from","my","your","very","one"]
df["clean_text"]=df['Tweets'].apply(lambda x: ' '.join([word for word in x.split() if word not in (word_stop)]))
#length of tweets using len()
df['len'] = df['Tweets'].str.len()

df
```

# Using Text Pattern (RE) + Language Processing

```python
def clean_text(list_of_words):
    word_stop=["is","are","you","so","a","in","up","the","as","it","I","can","the","did","an","You","it's","of","from","my","your","very","one"]
    clean_text = []
    for word in list_of_words:
        if word not in word_stop:
            clean_text.append(word)
    return " ".join(clean_text)

url = "https://confrecordings.ams3.digitaloceanspaces.com/Reviews.csv"
df2 = pd.read_csv(url)

df2["words"] = df2["Tweets"].apply(lambda x: re.findall("[a-zA-Z]+", x.lower()))

df2["clean_text"] = df2["words"].apply(lambda x: clean_text(x))

df2["clean_len"] = df2["clean_text"].apply(lambda x: len(x))
df2["original_len"] = df2["Tweets"].apply(lambda x: len(x))
df2
```
